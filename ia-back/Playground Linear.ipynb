```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Création du DataFrame avec les données fournies

df = pd.read_csv('./data/data.csv', sep=';')

# Séparation des variables indépendantes (X) et de la variable dépendante (y)
X = df[['surface']]
y = df['price']

# Création du modèle de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Prédictions sur les mêmes données
y_pred = model.predict(X)

# Visualisation de la régression linéaire
plt.scatter(X, y, color='blue', label='Données réelles')
plt.plot(X, y_pred, color='red', label='Régression linéaire')
plt.title('Régression linéaire entre la surface et le prix')
plt.xlabel('Surface (m²)')
plt.ylabel('Prix (€)')
plt.legend()
plt.grid(True)
plt.show()

# Affichage des coefficients de la régression
coef = model.coef_
intercept = model.intercept_

# Affichage des coefficients de la régression
print("Coefficient :", model.coef_)
print("Intercept :", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Calculer le MSE, RMSE et R²
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Afficher les résultats dans la console
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}")

# Visualisation des métriques dans un graphique
metrics = ['MSE', 'RMSE', 'R²']
values = [mse, rmse, r2]

plt.bar(metrics, values, color=['blue', 'orange', 'green'])
plt.title('Métriques de performance du modèle')
plt.ylabel('Valeurs')
plt.grid(True)
plt.show()
```

la régression linéaire peut tout à fait prendre en compte plusieurs variables explicatives (caractéristiques), comme la surface, le nombre de chambres, et le nombre de fenêtres. C'est ce qu'on appelle une régression linéaire multiple, où plusieurs variables indépendantes sont utilisées pour prédire une variable dépendante (ici, le prix d'un appartement).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('./data/data.csv', sep=';')

# Séparation des variables indépendantes (X) et de la variable dépendante (y)
# On utilise plusieurs variables explicatives
X = df[['surface', 'nbRooms', 'nbWindows']]
y = df['price']  # Variable cible

# Création du modèle de régression linéaire
model = LinearRegression()
model.fit(X, y)

# Prédictions sur les mêmes données
y_pred = model.predict(X)

# Calculer le MSE, RMSE et R²
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Afficher les résultats dans la console
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R²: {r2}")

# Affichage des coefficients de la régression
print("Coefficients :", model.coef_)
print("Intercept :", model.intercept_)

# Visualisation des métriques dans un graphique
metrics = ['MSE', 'RMSE', 'R²']
values = [mse, rmse, r2]

plt.bar(metrics, values, color=['blue', 'orange', 'green'])
plt.title('Métriques de performance du modèle')
plt.ylabel('Valeurs')
plt.grid(True)
plt.show()
```
